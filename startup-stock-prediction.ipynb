```python
#import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm

#getting dataset
dataset = pd.read_csv('/content/drive/MyDrive/startup.csv')
print(dataset.head())
print(dataset.tail())

#ci is confidence internal, None shows no confiedence level will be plotted
#lmplotr is an lineplot
sns.lmplot( x="R&D Spend", y="Profit", data=dataset, ci=None)
plt.show()

sns.barplot( x="R&D Spend", y= "Profit", data= dataset,width = 0.5)
plt.show()

#reshaping the value to 2 dim array
x = np.array(dataset['R&D Spend']).reshape(-1,1)
y= np.array(dataset['Profit']).reshape(-1,1)

#train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)

regr = LinearRegression()
#dim = x_train, x_test, y_train, y_test.ndim
#print(dim)

#fit of lin reg in to x and y
regr.fit(x_train, y_train)

#finding of Regr score
print(regr.score(x_train, y_train))


#predict y using x
y_pred = regr.predict(x_test)

#plot to show the regression lines
plt.scatter(x_test, y_test, color='green')
plt.plot(x_test, y_pred, color='orange')

plt.show()

#R2 score print( make the value near to 1)
print("R2 score=", round(sm.r2_score(x_test, y_test), 2))
```
